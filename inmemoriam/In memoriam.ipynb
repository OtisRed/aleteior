## Lewiatan

W historii myśli politycznej jedną z najbardziej znanych metafor państwa jest wymyślona przez Thomasa Hobbesa  __figura Lewiatana__, ogromnego potwora gotowego pożreć wszystko, co stoi mu na drodze. Potworem jest państwo, a jego żarłem - obszary wolności, w których porusza(ł) się poddany Lewiatana. A ściślej: jedynym słusznym wyborem poddanego jest poddanie się władzy Lewiatana.


![ziomek_lewiatan](pics/good_guy_leviatan.jpg)

<center>Tytułowy potwór na okładce <i>Lewiatana</i> ofiarowanej królowi Karolowi II. 1651
Źródło: https://commons.wikimedia.org/wiki/File:Drawing_of_frontispiece_of_Leviathan.jpg</center>


Może Lewiatan kiedyś był tylko efektowną metaforą, działającą na wyobraźnie studentów ówczesnych uniwersytetów. Dzisiaj jednak obserwując zapędy ustawodawcy, można stwierdzić, że omawiana przenośnia stała się rzeczywistością: w zbiurokratyzowanych czasach widzimy, jak machina państwowa wywłaszcza z prywatnej sfery coraz więcej spraw i zaczyna je regulować milionem przepisów. 

Tyczy się to oczywiście podatków i danin, które obciążają wielość zdarzeń – od sprzedaży chleba przez zbycie samochodu po zjedzenie schabowego na rodzinnym pikniku organizowanym przez Twojego szefa. 

Lewiatan pożera coraz więcej zdarzeń prawnych. Nie mogąc ich wszystkich przetrawić, zaczyna cierpieć w pewnym momencie na niestrawność – jego układ trawienny ma swoje moce przerobowe. Wiele mówi się o nadprodukcji prawa przez Panie i Panów z Sejmu - ale nie o tej przypadłości mowa.

Prawdziwa niestrawność dotyka obszaru stosowania prawa, a w szczególności jego wykładni (interpretacji). Nasz polski Lewiatan cierpi na __biegunkę interpretacyjną__ – a jej przejawem jest prawie __350.000 interpretacji podatkowych__ wydanych na przestrzeni ostatnich kilkunastu lat. Interpretacje podatkowe są właśnie symptomem pewnej choroby, ale w dosyć pokrętnym sensie. Miały bowiem stanowić lekarstwo na niejednoznaczność przepisów i wypływającą z niej samowolę urzędników w interpretowaniu prawa podatkowego.

I tak jak liczba sprzedawanych leków świadczy o zwiększonym odsetku chorych ludzi, tak i liczba interpretacji indywidualnych świadczy o problemach społeczeństwa z prawem podatkowym. 


## Interpretacje podatkowe - krótki wstęp 

Indywidualna interpretacja przepisów prawa podatkowego („interpretacja podatkowa”, "interpretacja", "interpretacja indywidualna") jest wydawana przez organ podatkowy na wniosek podatnika i chroni go w następujący sposób:<br><br>


<p><center><i>Jeśli <b>podatnik będzie postępował dokładnie tak, jak organ określi w wydanej dla podatnika interpretacji</b> – w razie przyszłej ewentualnej kontroli podatkowej, <b>organy państwa nie będą mogły zinterpretować prawa w inny sposób, jak określony w interpretacji</b>. </i></center></p>

Podatnik jest obowiązany we wniosku o wydanie interpretacji do pełnej prezentacji stanu faktycznego (zdarzenia przyszłego) i zadaje pytanie prawne. Przedstawia również  swoje stanowisko (interpretację), jakie przepisy i w jaki sposób powinny zadziałać w jego sytuacji.

Interpretacja indywidualna przypomina klasyczny model stosowania prawa (subsumpcja), co przedstawiłem poniżej w tabelce (przypomina, ponieważ subsumpcja dotyczy tylko bardziej wiążących aktów prawnych, jak decyzje czy wyroki). 

<center><i> Przykładowy problem i przykładowa interpretacja</i></center>

Krok|Model | Przykład
:-:|:-:|:-:
1|Organ ma ustalić, jakie normy prawne powinny obowiązywać w sytuacji opisanej przez podatnika. | Polski podatnik płaci podatek od darowizny. Jeśli podatnik uzyskuje darowiznę od rodzica, to darowizna nie podlega opodatkowaniu do 9637 PLN. 
2|Następnie musi ująć dany stan faktyczny w pojęciach normy („nazwać” poszczególne elementy fachowymi nazwami wziętymi wprost z przepisu prawnego). | Mama Kowalskiego [Rodzic] dała Kowalskiemu [polskiemu podatnikowi]  7000 PLN [darowiznę w wysokości 7000 PLN]. 
3|Po analizie dokonanej w Kroku 1 i 2 organ może wreszcie stwierdzić, że stan faktyczny rzeczywiście podlega (nie podlega) normie prawnej. | Kowalski nie podlega opodatkowaniu. 
4|I na samym końcu stwierdzić, jakie konsekwencje prawne wiążą się z podleganiem / niepodleganiem normie prawnej.  | Kowalski nie płaci podatku. 

Podatnik po uzyskaniu interpretacji ma trzy wyjścia:
1. zastosować się do niej – wtedy jego działania są chronione interpretacją, 
2. nie zastosować się; w razie ewentualnej kontroli dalej może spierać się z organem i dalej może kwestionować zdanie organu przed sądem;
3. zaskarżyć negatywną interpretację przed sądem administracyjnym.

Trzeba też wspomnieć, że bardzo istotny jest __rzetelny opis sytuacji (teraźniejszej lub przyszłej) będącej przedmiotem interpretacji__. Jeśli podatnik nie poda w wyczerpujący sposób wszystkich elementów stanu faktycznego (zdarzenia przyszłego), wobec której mają zostać zastosowane normy prawa podatkowego, __jedyną sankcją__, jaka go spotka, będzie w zasadzie brak możliwości skorzystania z „zabezpieczającego” charakteru interpretacji podatkowej.

Schemat wygląda przejrzyście w tabelce, ale w rzeczywistości sam proces wykładni prawa jest __skomplikowany__ i nie tylko dlatego, że składa się na nią zespół reguł wypracowanych przez mecenasów i profesorów prawa: sądy i organy dowolnie wykorzystują te niepisane reguły, czego przykładów dostarcza [prasa](https://www.money.pl/gospodarka/wiadomosci/artykul/podatkowy-absurd-opinia-skarbowki-podatki,95,0,2400351.html). 

W tym niewielkim eseju pokażę, w jaki sposób można uchwycić tendencje zmian w tematyce interpretacji na przestrzeni lat i jak jedna zmiana prawna mogła (prawdopodobnie) odstraszyć podatników od wnioskowania o tę swoistą tarczę podatkową.

## GAAR a sprawa PIT-owska

Jak widać większość interpretacji dotyczy VAT, co raczej nie powinno dziwić: w __zasadzie każda transakcja jest ovatowana__, a obowiązek zapłaty VAT jest jednym z najbardziej uciążliwych i czasochłonnych (comiesięczne deklaracje). Przedsiębiorcy chcą upewnić się, jaka stawka podatku obejmuje świadczone przez nich usługi lub sprzedawane towary. 

Drugie w klasyfikacji pod względem liczebności są podatki dochodowe __PIT (od osób fizycznych) oraz CIT (od osób prawnych)__, czyli podatki, które podatnicy płacą raz w roku (co do zasady). 

Widać również wzrastającą tendencję pomiędzy latami 2009 i 2015 oraz nagłe jej załamanie, którego pierwszym rokiem jest 2016 (zmniejszenie liczby wydawanych interpretacji PIT-owskich i CIT-owskich), a tendencja spadkowa trwa do 2017&nbspr. 

Dlaczego podatnicy przestali wnioskować o interpretacje w 2016 r.? 

Nie będziemy czytać wszystkich interpretacji, aby spróbować odpowiedzieć na to pytanie. Można dokonać analizy najczęściej występujących słów w interpretacjach PIT-owskich.

Teksty poszczególnych interpretacji podzieliłem na tokeny, a następnie na lematy za pomocą biblioteki Morfeusz (o czym __możesz poczytać__[tutaj](https://github.com/ArturTan/aleteior/blob/master/inmemoriam/PS.ipynb)). Dla potrzeb niniejszego tekstu przez lemat rozumiem najbardziej podstawową formę danego wyrazu (więcej o tokenach [tutaj](https://en.wikipedia.org/wiki/Lexical_analysis), a o lematach [tutaj](http://sgjp.pl/morfeusz/morfeusz.html.po)).

W analizie będziemy się starać wyeliminować najczęściej występujące słowa, które nie będą niosły ze sobą zbyt dużego ładunku znaczeniowego (takie jak spójniki, niektóre często występujące czasowniki). __Usuniemy__ również te słowa, które są charakterystyczne dla samego języka prawniczego / urzędniczego używanego w interpretacjach o PIT ("interpretacja", "podatek", "prawo", "prawny", "dochodowy") oraz pewne nalecialości, które nie zostały wyłapane przez algorytm Morfeusza. 

Do agregacji użyjemy algorytmu TF-IDF (ang. <i>term frequency-inverse document frequency</i>). Pozwala on na osiągnięcie bardzo ważnej rzeczy:  możemy wyodrębnić z danego dokumentu takie słowa, które świadczą o swoistości (indywidualności) tego dokumentu __względem__ całego korpusu dokumentów (przeczytaj o tym [tutaj](http://scikit-learn.org/stable/modules/feature_extraction.html#text-feature-extraction)). 

Działa to w następujący sposób:
1. liczymy częstotliwość występowania słowa W dla danego dokumentu [TF(W)], 
2. liczymy częstotliwość występowania słowa W dla całego korpusu - [DF(W)] i "odwracamy" ją  według wzoru:

   \\(IDF(W) = \log\frac{1 + n_d}{1 + DF(W)} + 1\\)
     
   gdzie \\(n_d\\) oznacza liczbę wszystkich dokumentów w korpusie. 
   
3. waga dla danego słowa w danym dokumencie wynosi TF(W) x IDF(W), przy czym jest ona dodatkowo normalizowana po uzyskaniu wszystkich wartości. 



Powyższy algorytm zastosujemy odnośnie trzech lat (2015, 2016, 2017) dla interpretacji indywidualnych z zakresu PIT. Następnie wybierzemy dla każdej interpretacji 100 słów najbardziej dla niej charakterystycznych (czyli 100 słów z największą wartością TF-IDF). Na samym końcu policzymy, jak często dane słowa są najbardziej charakterystyczne dla interpretacji wydanych w konkretnym roku.
